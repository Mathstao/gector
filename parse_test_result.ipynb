```python
import json
import Levenshtein
import pandas as pd
from gector.spell_checker import SymSpellChecker

spell_cheker = SymSpellChecker()
```

```python
rows = []

for iterations in [2, 3, 4]:
    for min_probability in [0.5, 0.6, 0.7, 0.8, 0.9]:
        for min_error_probability in [0.7, 0.8, 0.9]:
            index = "{}_{}_{}_False.json".format(iterations, min_probability, min_error_probability)
            result = json.load(open("./result/sampled_test_result_{}".format(index)))
            for cased_type in ['cased', 'uncased']:
                time_costs = []
                keep_correct_sent = 0
                change_correct_sent = 0
                full_correct_wrong_sent = 0
                part_correct_wrong_sent = 0
                not_correct_wrong_sent = 0

                for (source, target, cor, correction, ts) in result:
                    if cased_type == 'uncased':
                        source = source.lower()
                        target = target.lower()
                        cor= cor.lower()
                    
                    time_costs.append(ts)
                    if source==target:
                        if cor==target:
                            keep_correct_sent += 1
                        else:
                            change_correct_sent += 1
                    else:
                        if cor==target:
                            full_correct_wrong_sent += 1
                        else:
    #                         old_dist = Levenshtein.distance(source, target)
    #                         new_dist = Levenshtein.distance(cor, target)
    #                         if new_dist < old_dist:
                            if cor != source:
    #                             print(source)
    #                             print(cor)
    #                             print(target)
    #                             print()
                                part_correct_wrong_sent += 1
                            else:
                                not_correct_wrong_sent += 1
    #                             print(source)
    #                             print(cor)
    #                             print(target)
    #                             print()

                print(index)
                print(cased_type)
                print('keep_correct_sent'.ljust(25), keep_correct_sent)
                print('change_correct_sent'.ljust(25), change_correct_sent)
                print('full_correct_wrong_sent'.ljust(25), full_correct_wrong_sent)
                print('part_correct_wrong_sent'.ljust(25), part_correct_wrong_sent)
                print('not_correct_wrong_sent'.ljust(25), not_correct_wrong_sent)
                average_time_cost = str(round(sum(time_costs)/len(time_costs)*1000, 2)) + 'ms'
                print(average_time_cost)
                print()

                row = {
                    'iterations': iterations,
                    'min_probability': min_probability,
                    'min_error_probability': min_error_probability,
                    # 'add_spell_check': False,
                    'cased_type': cased_type,
                    'keep_correct_sent': keep_correct_sent,
                    'change_correct_sent': change_correct_sent,
                    'full_correct_wrong_sent': full_correct_wrong_sent,
                    'part_correct_wrong_sent': part_correct_wrong_sent,
                    'not_correct_wrong_sent': not_correct_wrong_sent,
                    'average time cost': average_time_cost,
                }
                rows.append(row)
```

```python
df = pd.DataFrame(rows)
df
```
